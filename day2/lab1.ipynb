```r

# Load the data.table library
library(data.table)

# Read the CitiBike trip data for September 2018
df <- fread("2018-citibike-tripdata/9_September/201809-citibike-tripdata_1.csv")

df=data.table(df)
# Show first few rows
head(df)
```

```r
# View the first 10 lines
cat("=== FIRST 10 LINES ===\n")
head(df, 10)

cat("\n\n=== FIRST 20 LINES ===\n")
# View the first 20 lines
head(df, 20)
```

```r
cat("=== TRIP DURATION STATISTICS ===\n\n")

# Calculate all statistics
trip_mean <- mean(df$tripduration, na.rm = TRUE)
trip_median <- median(df$tripduration, na.rm = TRUE)
trip_sd <- sd(df$tripduration, na.rm = TRUE)
trip_iqr <- IQR(df$tripduration, na.rm = TRUE)

# Display each statistic separately
cat("Mean:", round(trip_mean, 2), "seconds\n")
cat("Median:", round(trip_median, 2), "seconds\n")
cat("Standard Deviation:", round(trip_sd, 2), "seconds\n")
cat("IQR:", round(trip_iqr, 2), "seconds\n")

cat("\n--- Converted to minutes ---\n")
cat("Mean:", round(trip_mean/60, 2), "minutes\n")
cat("Median:", round(trip_median/60, 2), "minutes\n")
cat("Standard Deviation:", round(trip_sd/60, 2), "minutes\n")
cat("IQR:", round(trip_iqr/60, 2), "minutes\n")

```

```r
# TASK 5a: Find the 67th percentile and 4th decile of tripduration
cat("=== PERCENTILES AND DECILES ===\n\n")

# Calculate 67th percentile
percentile_67 <- quantile(df$tripduration, probs = 0.67, na.rm = TRUE)

# Calculate 4th decile (40th percentile)
decile_4 <- quantile(df$tripduration, probs = 0.4, na.rm = TRUE)

# Display results
cat("67th percentile:", round(percentile_67, 2), "seconds (", round(percentile_67/60, 2), "minutes)\n")
cat("4th decile (40th percentile):", round(decile_4, 2), "seconds (", round(decile_4/60, 2), "minutes)\n")

# Definitions
cat("\n=== DEFINITIONS ===\n")
cat("• 67th PERCENTILE: The value below which 67% of the trip durations fall.\n")
cat("  This means 67% of trips are shorter than", round(percentile_67/60, 2), "minutes.\n\n")
cat("• 4th DECILE (40th percentile): The value below which 40% of the trip durations fall.\n")
cat("  This means 40% of trips are shorter than", round(decile_4/60, 2), "minutes.\n")
cat("  Deciles divide data into 10 equal parts (1st decile = 10th percentile, 2nd = 20th, etc.)\n")
```

```r
# TASK 5b: Find total number of trips and breakdown by gender
cat("=== TRIP COUNTS ===\n\n")

# Total number of trips (rows in the dataset)
total_trips <- nrow(df)
cat("Total number of trips:", total_trips, "\n\n")

cat("Gender distribution:\n")

# Count trips by gender (1 = men, 2 = women)
gender_counts <- table(df$gender, useNA = "ifany")
print(gender_counts)

# More detailed breakdown
men_trips <- sum(df$gender == 1, na.rm = TRUE)
women_trips <- sum(df$gender == 2, na.rm = TRUE)
unknown_gender <- sum(is.na(df$gender) | df$gender == 0)

cat("\nDetailed breakdown:\n")
cat("Men (gender = 1):", men_trips, "trips\n")
cat("Women (gender = 2):", women_trips, "trips\n")
cat("Unknown/Other gender:", unknown_gender, "trips\n")

# Percentages
cat("\nPercentages:\n")
cat("Men:", round(men_trips/total_trips * 100, 1), "%\n")
cat("Women:", round(women_trips/total_trips * 100, 1), "%\n")
cat("Unknown/Other:", round(unknown_gender/total_trips * 100, 1), "%\n")
```

```r
# TASK 6: Make a histogram for birth year column
cat("=== HISTOGRAM OF BIRTH YEAR ===\n\n")

if(!is.null(df$`birth year`)) {
  # Clean the data
  birth_years <- df$`birth year`
  birth_years_clean <- birth_years[!is.na(birth_years) & birth_years > 1900 & birth_years < 2020] 
}

cat("Valid birth years:", length(birth_years_clean), "out of", length(birth_years), "\n")
cat("Birth year range:", min(birth_years_clean), "to", max(birth_years_clean), "\n\n")

# Create histogram
hist(birth_years_clean, 
      main = "Distribution of Birth Years - CitiBike Users",
      xlab = "Birth Year",
      ylab = "Frequency (Number of Users)",
      col = "lightblue",
      border = "darkblue",
      breaks = 20)

# Add some statistics to the plot
abline(v = mean(birth_years_clean), col = "red", lwd = 2, lty = 2)
abline(v = median(birth_years_clean), col = "green", lwd = 2, lty = 2)
legend("topleft", 
        legend = c("Mean", "Median"), 
        col = c("red", "green"), 
        lty = 2, lwd = 2)


cat("\n=== WHAT IS A HISTOGRAM? ===\n")
cat("A histogram is a graphical representation that shows the distribution of a dataset.\n")
cat("It divides the data into bins (intervals) and shows how many data points fall into each bin.\n")
cat("Key features:\n")
cat("• X-axis: Shows the range of values (birth years)\n")
cat("• Y-axis: Shows frequency (how many people born in each year range)\n")
cat("• Bars: Height represents the number of observations in each bin\n")
cat("• Shape: Reveals patterns like normal distribution, skewness, or multiple peaks\n")
```

```r
# TASK 7: Make a boxplot for birth year
cat("=== BOXPLOT OF BIRTH YEAR ===\n\n")


if(!is.null(df$`birth year`)) {
  # Create boxplot
  boxplot(birth_years_clean,
          main = "Boxplot of Birth Years - CitiBike Users",
          ylab = "Birth Year",
          col = "lightcoral",
          border = "darkred",
          horizontal = FALSE)
  
  # Add grid for easier reading
  grid(nx = NA, ny = NULL, col = "gray", lty = "dotted")
  
  # Calculate and display key statistics
  q1 <- quantile(birth_years_clean, 0.25)
  median_year <- median(birth_years_clean)
  q3 <- quantile(birth_years_clean, 0.75)
  iqr <- q3 - q1
  
  cat("Key statistics from the boxplot:\n")
  cat("Q1 (25th percentile):", q1, "\n")
  cat("Median (50th percentile):", median_year, "\n")
  cat("Q3 (75th percentile):", q3, "\n")
  cat("IQR (Interquartile Range):", iqr, "\n")
  
} else {
  cat("Birth year column not found. Available columns:\n")
  print(names(df))
}

cat("\n=== WHAT IS A BOXPLOT? ===\n")
cat("A boxplot (box-and-whisker plot) is a standardized way to display data distribution.\n")
cat("It shows five key statistics:\n\n")
cat("• BOTTOM WHISKER: Minimum value (or 1.5×IQR below Q1)\n")
cat("• BOTTOM OF BOX (Q1): 25th percentile - 25% of data is below this line\n")
cat("• MIDDLE LINE (MEDIAN): 50th percentile - middle value that splits data in half\n")
cat("• TOP OF BOX (Q3): 75th percentile - 75% of data is below this line\n")
cat("• TOP WHISKER: Maximum value (or 1.5×IQR above Q3)\n")
cat("• DOTS (if any): Outliers - values that fall outside the whiskers\n\n")
cat("The BOX contains the middle 50% of the data (from Q1 to Q3).\n")
cat("Boxplots are excellent for:\n")
cat("• Comparing distributions between groups\n")
cat("• Identifying outliers\n")
cat("• Understanding data spread and skewness\n")
```
